```python
# Using Regular Expressions (regex) we can scan through large files and extract words and values
import re
import numpy as np

def extract_asteroid_data(inserted_text):
    # 1. Read the entire file content
    with open(inserted_text, 'r') as text_file:
        all_content = text_file.read()

    # 2. Use the regex pattern to find all matches
    # In other words, the strong says:
    # Extract the date (skip AD), the extract x,y,z and also vx,vy,vz
    # Example line of the text file: 
        #2461159.500000000 = A.D. 2026-Apr-29 00:00:00.0000 TDB 
        # X = 4.099531219235758E+07 Y = 1.079738625661932E+08 Z =-5.939409462479845E+06
        # VX=-3.344164822966798E+01 VY=-4.689912147509263E+00 VZ= 3.194695911463352E+00
        # LT= 3.857571373439224E+02 RG= 1.156470803953781E+08 RR=-1.639741651676842E+01
    scanner = r"(\d+\.\d+) = A\.D\..*?\n\s+X\s?=\s?([\d\.E+-]+)\s+Y\s?=\s?([\d\.E+-]+)\s+Z\s?=\s?([\d\.E+-]+)\n\s+VX=\s?([\d\.E+-]+)\s+VY=\s?([\d\.E+-]+)\s+VZ=\s?([\d\.E+-]+)"
    matches = re.findall(scanner, all_content)

    # 3. Store results in a structured format (Dictionary or NumPy array)
    results = []
    for i in matches:
        data = {
            "Julian date": float(i[0]),
            "Vector": np.array([float(i[1]), float(i[2]), float(i[3]), 
                               float(i[4]), float(i[5]), float(i[6])])
        }
        results.append(data)
    
    return results

# Use and print the extracted values:
data_points = extract_asteroid_data("SSD_JPL_Horizon_TU3_orbit_data.txt")
print(f"Successfully extracted {len(data_points)} data points.")
```
